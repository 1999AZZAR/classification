## 13 Text Cleaning

```python
def txt_clnng(text, stem='Spacy'):

    output = ''

    # Remove HTML tags
    text = BeautifulSoup(text, 'html.parser').get_text()

    # Make lower
    text = text.lower()

    # Remove line breaks
    text = re.sub(r'\n', '', text)

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Remove stop words
    text = text.split()
    useless_words = nltk.corpus.stopwords.words('english')
    useless_words = useless_words + ['hi', 'im']

    text = [word for word in text if not word in useless_words]

    # Remove numbers
    text = [re.sub(r'\w*\d\w*', '', w) for w in text]

    # Remove emoticons
    text = [re.sub(r'[^\w\s]', '', w) for w in text]

    # Remove extra spaces
    text = [re.sub(r'\s+', ' ', w, flags=re.I) for w in text]

    # Stem or Lemmatize
    if stem == 'Stem':
        stemmer = PorterStemmer()
        text_stemmed = [stemmer.stem(y) for y in text]
    elif stem == 'Lem':
        lem = WordNetLemmatizer()
        text_stemmed = [lem.lemmatize(y) for y in text]
    elif stem == 'Spacy':
        text = nlp(' '.join(text))
        text_stemmed = [y.lemma_ for y in text]
    else:
        text_stemmed = text

    output = ' '.join(text_stemmed)

    return output
```

## 14 RoBERTa Labeling

```python
def lrbrt(text):
  ncdd = lTokenizer(text, return_tensors='pt')
  output = lModel(**ncdd)
  scores = output[0][0].detach().numpy()
  scores = softmax(scores)
  scores = {
      'RoBERTa_neg': scores[0],
      'RoBERTa_neu': scores[1],
      'RoBERTa_pos': scores[2]
  }
  plolarity = list(scores.values()).index(max(scores.values()))
  sentiment_mapping = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
  scores['RoBERTa_pol'] = sentiment_mapping[plolarity]
  return scores
```

```python
### Scoring text data
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df), desc='Scoring texts'):
    try:
      text = row['clean']
      id = row['Id']
      RoBERTa = lrbrt(text)
      res[id] = RoBERTa
    except RuntimeError:
      print(f'Error at {i}')
```

```python
df1 = pd.DataFrame(res).T.reset_index().rename(columns={'index': 'Id'}).merge(df, how='left')
```

# 3 Feature Extraction

```python
def mbdd(input: list[str], tokenizer: bertTokenizer, model: bertModel, max_length: int = 100, verbose:bool = True):
    mbddngs = []
    if verbose:
        iterator = tqdm(input, desc='Embedding texts')
    else:
        iterator = input

    for i, text in enumerate(iterator):
        inputs = bertTokenizer.encode_plus(text, add_special_tokens=True, padding=True, truncation=True, max_length=max_length, return_tensors='pt')
        with torch.no_grad():
            output = model(**inputs)
            embeddings = output.last_hidden_state.mean(dim=1)
            mbddngs.append(embeddings)

    return torch.cat(mbddngs, dim=0)
```

## 42 Data Spliting

```python
def tvt_splt(X, y, split_shape=[70, 20, 10], random_state=42):
  tt = float(split_shape[2]/sum(split_shape))
  tv = float(split_shape[1]/sum(split_shape[:2]))

  if tt != 0:
    X_tt, X_test, y_tt, y_test = train_test_split(X.numpy(), y, test_size=tt, random_state=random_state)
  else:
    X_tt, y_tt = X.numpy(), y
    X_test, y_test = [], []

  X_train, X_val, y_train, y_val = train_test_split(X_tt, y_tt, test_size=tv, random_state=random_state)

  return {'X_train': X_train,'X_val': X_val, 'X_test': X_test, 'y_train': y_train, 'y_val': y_val, 'y_test': y_test}
```

## 51 RNN

```python
class accCallback(Callback):
    def __init__(self, threshold):
        super().__init__()
        self.threshold = threshold
    def on_epoch_end(self, epoch, logs=None):
        train_acc = logs.get('accuracy')
        val_acc = logs.get('val_accuracy')
        if train_acc >= self.threshold and val_acc >= self.threshold:
            self.model.stop_training = True

early_stopping = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=32, verbose=0, mode='auto', restore_best_weights=True,)

adam = Adam(learning_rate = ExponentialDecay(initial_learning_rate=5e-4, decay_steps=10000, decay_rate=1e-6))

huber = Huber(delta=0.5, reduction='sum_over_batch_size', name='huber_loss')
```

```python
model = Sequential([
    Input(X0_train.shape[1:]),
    Bidirectional(LSTM(128, return_sequences=True)),

    BatchNormalization(),
    Activation('relu'),

    # Dense(256, activation='relu'),
    # Dense(128, activation='relu'),
    # Dense(128, activation='relu'),

    Flatten(),
    Dense(128, activation='relu'),

    Dense(3, activation='softmax'),
    ])
```

```python
model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy', # huber,
    optimizer = adam,
    metrics=['accuracy']
    )
```

```python
from tensorflow.keras.optimizers import Adam

# Initialize the optimizer with a fixed learning rate
optimizer = Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint

# Learning rate scheduler to reduce LR on plateau
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

# Checkpoint to save the best model using the .keras format
checkpoint = ModelCheckpoint("best_model.keras", monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)

v721 = model.fit(
    X0_train,
    y0_train,
    epochs=256,
    batch_size=256,
    validation_data=(X0_val, y0_val),
    callbacks=[early_stopping, accCallback(threshold=0.92), lr_scheduler, checkpoint],
    shuffle=True
)
```

```python
train_acc = v721.history['accuracy']
val_acc = v721.history['val_accuracy']
train_loss = v721.history['loss']
val_loss = v721.history['val_loss']

plt.figure(figsize=(8, 3))

plt.subplot(1, 2, 1)
plt.plot(train_acc, label='Train', linestyle='-')
plt.plot(val_acc, label='Validation', linestyle='-')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_loss, label='Train', linestyle='--')
plt.plot(val_loss, label='Validation', linestyle='--')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()

plt.show()
```
