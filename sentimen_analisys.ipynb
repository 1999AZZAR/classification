## 1.3 Text Cleaning

```python
# Load Spacy Model
nlp = spacy.load('en_core_web_sm')

# Function to clean text
def txt_clnng(text, stem='Spacy'):
    output = ''
    text = BeautifulSoup(text, 'html.parser').get_text()
    text = text.lower()
    text = re.sub(r'\n', '', text)
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)
    text = text.split()
    useless_words = nltk.corpus.stopwords.words('english')
    useless_words = useless_words + ['hi', 'im']
    text = [word for word in text if not word in useless_words]
    text = [re.sub(r'\w*\d\w*', '', w) for w in text]
    text = [re.sub(r'[^\w\s]', '', w) for w in text]
    text = [re.sub(r'\s+', ' ', w, flags=re.I) for w in text]

    # Stem or Lemmatize
    if stem == 'Stem':
        stemmer = PorterStemmer()
        text_stemmed = [stemmer.stem(y) for y in text]
    elif stem == 'Lem':
        lem = WordNetLemmatizer()
        text_stemmed = [lem.lemmatize(y) for y in text]
    elif stem == 'Spacy':
        text = nlp(' '.join(text))
        text_stemmed = [y.lemma_ for y in text]
    else:
        text_stemmed = text

    output = ' '.join(text_stemmed)
    return output

# Function to apply text cleaning using multithreading
def parallel_txt_cleaning(df, column_name, stem='Spacy'):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Create a list of futures
        futures = [executor.submit(txt_clnng, text, stem) for text in df[column_name]]
        # Collect the results
        results = [future.result() for future in concurrent.futures.as_completed(futures)]
    return results

# Apply the function to your dataframe
df['clean'] = parallel_txt_cleaning(df, 'content', 'Spacy')
```

```python
df = df[df['clean'] != ''].dropna().reset_index(drop=True)
df.insert(0, 'Id', df.index + 1)
```

## 1.4 RoBERTa Labeling

```python
# Load a pre-trained sentiment analysis model
model_name = "cardiffnlp/twitter-roberta-base-sentiment"
lTokenizer = AutoTokenizer.from_pretrained(model_name)
lModel = AutoModelForSequenceClassification.from_pretrained(model_name)
lModel.eval()  # Set the model to evaluation mode

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
lModel.to(device)

def lrbrt(text):
    ncdd = lTokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    ncdd = {k: v.to(device) for k, v in ncdd.items()}
    
    with torch.no_grad():
        output = lModel(**ncdd)
    
    scores = softmax(output.logits, dim=1).cpu().numpy()[0]
    
    scores_dict = {
        'RoBERTa_neg': float(scores[0]),
        'RoBERTa_neu': float(scores[1]),
        'RoBERTa_pos': float(scores[2])
    }
    
    plolarity = scores.argmax()
    sentiment_mapping = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    scores_dict['RoBERTa_pol'] = sentiment_mapping[plolarity]
    
    return scores_dict
```

### Scoring text data

```python
# Process in batches
batch_size = 32
res = {}

for i in tqdm(range(0, len(df), batch_size), desc='Scoring texts'):
    batch = df.iloc[i:i+batch_size]
    
    for _, row in batch.iterrows():
        try:
            text = row['clean']
            id = row['Id']
            RoBERTa = lrbrt(text)
            res[id] = RoBERTa
        except Exception as e:
            print(f'Error processing Id {id}: {str(e)}')
```

```python
df1 = pd.DataFrame(res).T.reset_index().rename(columns={'index': 'Id'}).merge(df, how='left')
```

```python
# Convert results to DataFrame
result_df = pd.DataFrame.from_dict(res, orient='index')

# Optimized label encoding
label_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
result_df['label'] = result_df['RoBERTa_pol'].map(label_map)

# If you need to keep the original DataFrame structure
df1 = df.join(result_df, on='Id')
```

# 3 Feature Extraction

```python
def mbdd(input: list[str], tokenizer: BertTokenizer, model: BertModel, max_length: int = 100, batch_size: int = 32, verbose: bool = True):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    embeddings = []
    
    for i in tqdm(range(0, len(input), batch_size), desc='Embedding texts', disable=not verbose):
        batch = input[i:i+batch_size]
        
        inputs = tokenizer(batch, add_special_tokens=True, padding=True, truncation=True, 
                           max_length=max_length, return_tensors='pt')
        inputs = {k: v.to(device) for k, v in inputs.items()}
        
        with torch.no_grad():
            output = model(**inputs)
            batch_embeddings = output.last_hidden_state.mean(dim=1)
            embeddings.append(batch_embeddings.cpu())

    return torch.cat(embeddings, dim=0)
```

## 4.2 Data Spliting

```python
def tvt_splt(X, y, split_shape=[70, 20, 10], random_state=42):
  tt = float(split_shape[2]/sum(split_shape))
  tv = float(split_shape[1]/sum(split_shape[:2]))

  if tt != 0:
    X_tt, X_test, y_tt, y_test = train_test_split(X.numpy(), y, test_size=tt, random_state=random_state)
  else:
    X_tt, y_tt = X.numpy(), y
    X_test, y_test = [], []

  X_train, X_val, y_train, y_val = train_test_split(X_tt, y_tt, test_size=tv, random_state=random_state)

  return {'X_train': X_train,'X_val': X_val, 'X_test': X_test, 'y_train': y_train, 'y_val': y_val, 'y_test': y_test}
```

```python
class accCallback(Callback):
    def __init__(self, threshold):
        super().__init__()
        self.threshold = threshold
    def on_epoch_end(self, epoch, logs=None):
        train_acc = logs.get('accuracy')
        val_acc = logs.get('val_accuracy')
        if train_acc >= self.threshold and val_acc >= self.threshold:
            self.model.stop_training = True
```

```python
model = Sequential([
    Input(X0_train.shape[1:]),
    Bidirectional(LSTM(128, return_sequences=True)),

    BatchNormalization(),
    Activation('relu'),

    # Dense(256, activation='relu'),
    # Dense(128, activation='relu'),
    # Dense(128, activation='relu'),

    Flatten(),
    Dense(128, activation='relu'),

    Dense(3, activation='softmax'),
    ])

model.summary()
```
